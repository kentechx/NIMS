# Homo data from paper NIMS

```python
import igraph
import random
import matplotlib.pyplot as plt
```

```python
g = igraph.read("./src/exa_homo.lgl").as_undirected()
```

## Optimal Modularity

```python
def normalize(metrics, l=5., h=10.):
    # normalize to [5, 10]
    _min = min(metrics)
    _max = max(metrics)
    return [(m-_min)/(_max-_min)*(h-l)+l for m in metrics]
```

```python
cluster = g.community_optimal_modularity()
```

```python
plot, ax = plt.subplots(figsize=(10, 10))
ax.set_title("a")
ax.axis('off')
```

```python
# plot modules
%matplotlib inline
random.seed(81)
igraph.plot(cluster,
            "./src/modules.png",
            layout=g.layout_fruchterman_reingold(),
            mark_groups=True, vertex_label=g.vs['name'], margin=(100, 100, 100, 100))
```

```python
# plot with degree
metrics = g.degree()
random.seed(81)
print('range: [%f, %f]'%(min(metrics), max(metrics)))
metrics = normalize(metrics, 1, 20)
igraph.plot(cluster,
            './src/degree.png',
            layout=g.layout_fruchterman_reingold(),
            mark_groups=True, vertex_label=g.vs['name'], margin=(100, 100, 100, 100),
           vertex_size=metrics)
```

```python
# plot with betweenness
metrics = g.betweenness()
random.seed(81)
print('range: [%f, %f]'%(min(metrics), max(metrics)))
metrics = normalize(metrics, 1, 20)
igraph.plot(cluster,
            './src/betweenness.png',
            layout=g.layout_fruchterman_reingold(),
            mark_groups=True, vertex_label=g.vs['name'], margin=(100, 100, 100, 100),
           vertex_size=metrics)
```

```python
# plot with closeness
#fig, ax = plt.subplots()
metrics = g.closeness()
random.seed(81)
print('range: [%f, %f]'%(min(metrics), max(metrics)))
metrics = normalize(metrics, 1, 20)
igraph.plot(cluster,
            './src/closeness.png',
            layout=g.layout_fruchterman_reingold(),
            mark_groups=True, vertex_label=g.vs['name'], margin=(100, 100, 100, 100),
            vertex_size=metrics)
```

```python
# plot with clustering coefficient
metrics = g.transitivity_local_undirected()
random.seed(81)
print('range: [%f, %f]'%(min(metrics), max(metrics)))
metrics = normalize(metrics, 1, 20)
igraph.plot(cluster,
            './src/clustering_coef.png',
            layout=g.layout_fruchterman_reingold(),
            mark_groups=True, vertex_label=g.vs['name'], margin=(100, 100, 100, 100),
           vertex_size=metrics)
```

```python
# plot with eigenvector centrality
metrics = g.eigenvector_centrality()
random.seed(81)
print('range: [%f, %f]'%(min(metrics), max(metrics)))
metrics = normalize(metrics, 1, 20)
igraph.plot(cluster,
            './src/eigenvector.png',
            layout=g.layout_fruchterman_reingold(),
            mark_groups=True, vertex_label=g.vs['name'], margin=(100, 100, 100, 100),
           vertex_size=metrics)
```
